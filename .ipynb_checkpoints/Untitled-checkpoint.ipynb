```python
# 대신증권 연결 확인
import win32com.client

instCpCybos = win32com.client.Dispatch("CpUtil.CpCybos")
print(instCpCybos.IsConnect)

import pandas as pd
import numpy as np
import math
import time
from datetime import datetime,timedelta


pd.options.display.float_format = '{:.6f}'.format
```

```python
def get_code(name): 
    import win32com.client

    instCpCodeMgr = win32com.client.Dispatcsh("CpUtil.CpCodeMgr")
    codeList = instCpCodeMgr.GetStockListByMarket(1)

#     name_to_code = {}
    code_to_name = {}
    for code in codeList:
        name = instCpCodeMgr.CodeToName(code)
#         name_to_code[name] = code
        code_to_name[code] = name
    
    return code_to_name


# 분 차트 받아오기
def get_bidask(code):  # 종목, 기간, 오늘, 시점, 분, 시간간격
    dfs = []
    
    time_= 905
    
    while time_ > 899:
        
        if len(str(time_)) == 3:
            time_ = '0'+str(time_)
            
        instStockChart = win32com.client.Dispatch("Dscbo1.StockBid")
        instStockChart.SetInputValue(0, code)
        instStockChart.SetInputValue(2, 80)
        instStockChart.SetInputValue(3, ord('H'))
        instStockChart.SetInputValue(4, str(time_))
        instStockChart.BlockRequest()

        numData = instStockChart.GetHeaderValue(2)

        temp = {}
        for i in range(numData):
            temp[instStockChart.GetDataValue(9,i)] = [instStockChart.GetDataValue(2,i),instStockChart.GetDataValue(3,i) ]
        df = pd.DataFrame(temp).transpose()
        df.index.names = ['time']
        df.columns = ['bid','ask']
        dfs.append(df)

        time_ = int(time_) - 1
        
        time.sleep(0.3)
        print(time_)
        
    final = pd.concat(dfs)
    final.to.pickle('')
```
